## What Success Looks Like

A successful proposal will:

| Criterion | Expectation |
|-----------|-------------|
| **Clarity** | Problem is clearly stated and understandable |
| **Relevance** | Problem connects to chemical engineering |
| **Feasibility** | Data is accessible and sufficient for ML |
| **Scope** | Project is appropriately sized (not too big, not too small) |
| **Methods** | Proposed approaches are reasonable for the problem |
| **Metrics** | Clear criteria for evaluating success |

### Red Flags (things to avoid)

- Vague problem statements ("I want to use ML on chemical data")
- No data source identified
- Data with fewer than ~50 samples (unless justified)
- Methods that don't match the problem (e.g., classification for a regression task)
- No clear success criteria
